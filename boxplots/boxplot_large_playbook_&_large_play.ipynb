```python
import matplotlib.pyplot as plt
import numpy as np

# Create a box plot
plt.figure(figsize=(10, 6))
plt.boxplot(df['Lines of Code'])

# Get the statistical values
data = df['Lines of Code']
min_val = np.min(data)
max_val = np.max(data)
mean_val = np.mean(data)
q1 = np.percentile(data, 25)
median_val = np.percentile(data, 50)
q3 = np.percentile(data, 75)

# Highlight the statistical values

plt.axhline(y=min_val, color='red', linestyle='--', label='Min')
plt.axhline(y=max_val, color='blue', linestyle='--', label='Max')
plt.axhline(y=mean_val, color='green', linestyle='--', label='Mean')
plt.axhline(y=q1, color='orange', linestyle='--', label='Q1')
plt.axhline(y=median_val, color='purple', linestyle='--', label='Median')
plt.axhline(y=q3, color='black', linestyle='--', label='Q3')

# Set the x-axis labels
xtick_labels = ['Lines of Code']
plt.xticks(range(1, len(xtick_labels) + 1), xtick_labels)

# Set the y-axis limits
plt.ylim(0, 200)

# Set the plot title and axis labels
plt.title('Lines of Code')
plt.xlabel('Metrics')
plt.ylabel('Lines of Code')

# Add a legend
plt.legend()

# Display the plot
plt.tight_layout()
plt.show()
```

```python
def transformer_url(url):
    # Vérifier si l'URL est une chaîne
    if isinstance(url, str):
        # Remplacer le préfixe
        url_locale = url.replace("https://github.com", "C:\\Users\\PC\\Desktop\\mitacs\\Projet\\ansible\\playbooks\\projects")
        
        # Diviser l'URL en parties
        parties = url_locale.split('/')
        
        # Enlever 'blob', 'master', et 'main' une seule fois
        for mot in ['blob', 'master', 'main']:
            if mot in parties:
                parties.remove(mot)
        
        # Reconstituer l'URL avec des antislashs
        url_locale = "\\".join(parties)
        
        return url_locale
    else:
        return url  # Retourner l'objet tel quel s'il n'est pas une chaîne

df_manual['Playbook Path'] = df_manual['Playbook Path'].apply(transformer_url)
```

```python
# Sauvegarder le DataFrame modifié dans un nouveau fichier CSV
df_manual.to_csv('sample_local_paths.csv', index=False)
```

```python
import os

def extract_filename(path):
    if isinstance(path, str):
        base_name = os.path.basename(path)  # Récupère le nom du fichier avec son extension
        file_name, _ = os.path.splitext(base_name)  # Sépare le nom du fichier de son extension
        return file_name
    return None
```
